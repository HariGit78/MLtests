```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('big data proj').getOrCreate()
```

```python
from pyspark.sql.types import *
```

```python
df_br21 = spark.read.format("csv").option("header","True").load("/content/branch_data-dec2021.csv")
```

```python
df_br21.printSchema()
```

```python
df_br21.show()
```

```python
from pyspark.sql.functions import to_date, col
df_br21_1=df_br21.withColumn("bDATE",to_date(col("DATE"),"dd-MM-yyyy"))
```

```python
df_br21_1.printSchema()
```

```python
df_br21_1.count()
```

```python
df_br21_1.show(1,truncate=False)
```

```python
df_br21_2=df_br21_1.drop("DATE")
```

```python
df_br21_2.printSchema()
```

```python
df_br21_3=df_br21_2.withColumn("Current_Dep",col("CurrentDep").cast("Integer"))
```

```python
df_br21_3.printSchema()
```
