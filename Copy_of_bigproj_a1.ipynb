```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('big data proj').getOrCreate()
```

```python
from pyspark.sql.types import *
```

```python
df_br21 = spark.read.format("csv").option("header","True").load("/content/branch_data-dec2021.csv")
```

```python
df_br21.printSchema()
```

```python
df_br21.show()
```

```python
from pyspark.sql.functions import to_date, col
```

```python
df_br21=df_br21.withColumn("DATE",to_date(col("DATE"),"dd-MM-yyyy"))
df_br21=df_br21.withColumn("CurrentDep",col("CurrentDep").cast("Integer"))
df_br21=df_br21.withColumn("SavingsDep",col("SavingsDep").cast("Integer"))
df_br21=df_br21.withColumn("CASA",col("CASA").cast("Integer"))
df_br21=df_br21.withColumn("TermDep",col("TermDep").cast("Integer"))
df_br21=df_br21.withColumn("AggregateDep",col("AggregateDep").cast("Integer"))
df_br21=df_br21.withColumn("RetailCredit",col("RetailCredit").cast("Integer"))
df_br21=df_br21.withColumn("Agriculture",col("Agriculture").cast("Integer"))
df_br21=df_br21.withColumn("BusinessCredit",col("BusinessCredit").cast("Integer"))
df_br21=df_br21.withColumn("AggregateCredit",col("AggregateCredit").cast("Integer"))
df_br21=df_br21.withColumn("AggregateBusiness",col("AggregateBusiness").cast("Integer"))
```

```python
df_br21.printSchema()
```

```python
df_br21.count()
```

```python
df_br21.show(1,truncate=False)
```

```python
df_br21.registerTempTable("br1")
```

Print top 5 branches with CASA accounts in 2021

```python
q1=spark.sql("select branches,casa from (select branches, casa, dense_rank() over (order by casa desc) as rnk from br1 ) where rnk <=5 order by casa desc, branches")
```

```python
q1.show()
```

find total retail credit, agri credit and business credit accounts in 2021

```python
q2=spark.sql("select sum(retailcredit) as tot_retail_cr_2021, sum(agriculture) as tot_agri_cr_2021, sum(businesscredit) as tot_bus_cr_2021 from br1")
```

```python
q2.show()
```
