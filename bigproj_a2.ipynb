Find total Retail Credit, Agre Credit and business Credit accounts in 2021

```python
s21=y21.agg(F.sum("RetailCredit").alias("RetailCredit"),F.sum("Agriculture").alias("Agriculture"),F.sum("BusinessCredit").alias("BusinessCredit"))
```

```python
s21.show(truncate=False)
```

Find total Retail Credit, Agre Credit and business Credit accounts in 2022

```python
s22=y22.agg(F.sum("RetailCredit").alias("RetailCredit"),F.sum("Agriculture").alias("Agriculture"),F.sum("BusinessCredit").alias("BusinessCredit"))
```

```python
s22.show(truncate=False)
```

**Find the change in percentage of these accounts over previous years branch wise**

```python
# cross join the two single row data frames to have both years in one row
```

```python
both=s21.crossJoin(s22)
```

```python
# compute % change =((sum_22 - sum_21) / Nullif(sum_21,0))*100, rounded to 2
# Fix: Rename columns of s21 and s22 before joining to avoid ambiguous column names.
s21_renamed = s21.select(
    F.col("RetailCredit").alias("RetailCredit_21"),
    F.col("Agriculture").alias("Agriculture_21"),
    F.col("BusinessCredit").alias("BusinessCredit_21")
)
s22_renamed = s22.select(
    F.col("RetailCredit").alias("RetailCredit_22"),
    F.col("Agriculture").alias("Agriculture_22"),
    F.col("BusinessCredit").alias("BusinessCredit_22")
)

# Perform the cross join with the renamed DataFrames
both_renamed = s21_renamed.crossJoin(s22_renamed)

result=(both_renamed.select(
                F.round(
                    (F.col("RetailCredit_22")-F.col("RetailCredit_21")) /
								  F.when(F.col("RetailCredit_21")!=0, F.col("RetailCredit_21")) * 100.0,2
								  ).alias("retail_pct_change_22_vs_21"),
								F.round(
								(F.col("Agriculture_22") - F.col("Agriculture_21"))/
								F.when(F.col("Agriculture_21")!=0, F.col("Agriculture_21")) * 100.0,2
								).alias("agri_pct_change_22_vs_21"),
								F.round(
								(F.col("BusinessCredit_22") - F.col("BusinessCredit_21"))/
								F.when(F.col("BusinessCredit_21")!=0, F.col("BusinessCredit_21")) * 100.0,2
								).alias("business_pct_change_22_vs_21")
				)
)
```
